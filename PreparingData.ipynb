```python
import pandas as pd
```

```python
weather_raw = pd.read_csv("./data/HistoryDataWeatherToronto.csv")
```

```python
weather_raw.head(10)
```

```python
city_df = weather_raw.groupby("city_name")
toronto_full_df = city_df.get_group("Toronto")
```

```python
toronto_df = toronto_full_df[["dt_iso", "weather_id"]]
```

```python
merge_df = pd.merge(weather_raw, toronto_df, on='dt_iso', how='left')
```

```python
weather_sorted = merge_df.sort_values(by=['dt_iso'])
weather_sorted.head(100)
```

```python
weather_sorted[['dt_iso', 'city_name','weather_main_x', 'weather_main_y']]
```

```python
merge_df.to_csv('./data/WeatherData_TorontoLabel.csv', index=False ,sep=',', header=True, encoding='utf-8')
weather_sorted.to_csv('./data/SortedWeatherData_TorontoLabel.csv', index=False ,sep=',', header=True, encoding='utf-8')
```
