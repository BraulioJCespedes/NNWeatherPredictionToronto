```python
import pandas as pd
```

```python
weather_raw = pd.read_csv("./data/HistoryDataWeather.csv")
```

```python
weather_raw.head(10)
```

```python
city_df = weather_raw.groupby("city_name")
toronto_full_df = city_df.get_group("Toronto")
```

```python
toronto_full_df.loc[:, "dt_iso"] = pd.to_datetime(toronto_full_df["dt_iso"], format='%Y-%m-%d %H:%M:%S %z UTC', errors='coerce', utc=True)
```

```python
toronto_full_df.head(10)
```

```python
toronto_full_df['dt_iso'][0]
```

```python
toronto_full_df.loc[:, "dt_iso"] = toronto_full_df["dt_iso"].dt.date
```

```python
toronto_full_df.head(10)
```

```python
toronto_df = toronto_full_df[["dt_iso", "temp"]]
```

```python
toronto_df.to_csv('./data/toronto_temperature.csv', index=False ,sep=',', header=True, encoding='utf-8')
```

```python
merge_df = pd.merge(weather_raw, toronto_df, on='dt_iso', how='left')
merge_df = merge_df.drop(merge_df[merge_df["city_name"] == 'Toronto'].index)
```

```python
weather_sorted = merge_df.sort_values(by=['dt_iso'])
weather_sorted.head(100)
```

```python
weather_sorted[['dt_iso', 'city_name','weather_id_x', 'weather_id_y']]
```

```python
merge_df.to_csv('./data/WeatherData_TorontoLabel.csv', index=False ,sep=',', header=True, encoding='utf-8')
weather_sorted.to_csv('./data/SortedWeatherData_TorontoLabel.csv', index=False ,sep=',', header=True, encoding='utf-8')
```
